```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Fruit-Prices-2022.csv")
```

```python
# 1. Basic Graph (Scatter Plot)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="RetailPrice", y="CupEquivalentPrice", hue="Form", palette="viridis")
plt.title("Retail Price vs Cup Equivalent Price by Form")
plt.xlabel("Retail Price ($)")
plt.ylabel("Cup Equivalent Price ($)")
plt.legend(title="Form")
plt.tight_layout()
plt.show()
```

```python
#2. Bar Chart: Average Retail Price per Fruit

# Prepare data
avg_prices_df = df.groupby("Fruit", as_index=False)["RetailPrice"].mean().sort_values(by="RetailPrice", ascending=False)

# Set larger figure height for better spacing
plt.figure(figsize=(12, len(avg_prices_df) * 0.5))

# Warning-free bar plot with hue and no legend
sns.barplot(data=avg_prices_df, x="RetailPrice", y="Fruit", hue="Fruit", palette="coolwarm", dodge=False, legend=False)

plt.title("Average Retail Price per Fruit")
plt.xlabel("Average Retail Price ($)")
plt.ylabel("Fruit")
plt.tight_layout()
plt.show()
```

```python
# 3. Line Graph: Retail Price Trend (Sorted)

plt.figure(figsize=(10, 6))
sorted_df = df.sort_values("RetailPrice")
plt.plot(sorted_df["Fruit"], sorted_df["RetailPrice"], marker='o', linestyle='-', color='teal')
plt.xticks(rotation=90)
plt.title("Retail Price Trend per Fruit (Sorted)")
plt.xlabel("Fruit")
plt.ylabel("Retail Price ($)")
plt.tight_layout()
plt.show()
```

```python
# 4. Histogram: Distribution of Retail Prices

plt.figure(figsize=(8, 6))
sns.histplot(df["RetailPrice"], bins=10, kde=True, color="orange")
plt.title("Distribution of Retail Prices")
plt.xlabel("Retail Price ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# 5. Pie Chart: Proportion of Fruits by Form

plt.figure(figsize=(6, 6))
form_counts = df["Form"].value_counts()
plt.pie(form_counts, labels=form_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Fruit Forms")
plt.tight_layout()
plt.show()
```

```python
# 6. Box and Whiskers Chart: Retail Price by Form

plt.figure(figsize=(10, 6))

# Warning-free boxplot
sns.boxplot(data=df, x="Form", y="RetailPrice", hue="Form", palette="Set2", dodge=False, legend=False)

plt.title("Retail Price Distribution by Form")
plt.xlabel("Form")
plt.ylabel("Retail Price ($)")
plt.tight_layout()
plt.show()
```
